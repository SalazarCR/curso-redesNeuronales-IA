```python
import pandas as pd
```

Crear el dataframe

```python
training = pd.read_csv("train.csv")
training.head(5)
```

```python
training.info()
```

Verificar el contenido del dataframe training

```python
training['Gender'] = training['Gender'].apply(lambda toLabel:0 if toLabel=='male' else 1)
```

Reemplazar los datos del gГ©nero a entero
male==> 0           female==> 1

```python
training.head(5)
```

```python
training["Age"].fillna(training["Age"].mean(),inplace=True)
```

Completar los valores Nan correspondiente a la edad (Age) con el promedio de edades validos

```python
training.head(5)
```

```python
training.info()
```

Verificar los datos del dataset

```python
y_target = training["Survived"].values
print(y_target)
```

Identificar los input data y los target data

```python
columns = ["Fare","Pclass","Gender","Age","SibSp"]
x_input = training[list(columns)].values
print(x_input)
```

```python
import keras
from keras import layers
from keras import ops
```

Definir la estructura de la red neuronal para el aprendizaje

```python
model = keras.Sequential()
model.add(layers.Dense(32,input_dim=5,activation='relu'))
model.add(layers.Dense(32, activation="relu", name="layer1"))
model.add(layers.Dense(1, activation='sigmoid',name="layer2"))
```

```python
import numpy as np
XX = np.array(x_input[[10]])
print(XX)
respuesta = model.predict(XX)
print(respuesta.round()[0][0])
```

Configurar los parГЎmetros de la red

```python
model.fit(x_input,y_target,epochs=400)
```

Realizar el proceso de entrenamiento de la red neuronal

```python
score = model.evaluate(x_input,y_target)
print("\n %s: %.2f%%" % (model.metrics_names[1],score[1]*100))
```

Evaluar la precisiГіn.

```python
y_simulado = []
for i in range(len(y_target)):
  y_simulado.append(model.predict(x_input[[i]]).round())
print(y_simulado)
```

```python
respuesta = model.predict(np.array([[83.475, 1, 1, 35, 1,]]))
print(respuesta.round()[0][0])
```

```python
precision = 0
for i in range(len(y_simulado)):
  if y_target[i]==y_simulado[i].round()[0][0]:
    precision = precision + 1
precision = precision/float(len(y_simulado))
print(precision)
```

```python
y_simulado_lista = []
for i in range(len(y_simulado)):
  y_simulado_lista.append(y_simulado[i].round()[0][0])
print(y_simulado_lista)
```

```python
from sklearn.metrics import confusion_matrix
from matplotlib import pyplot as plt
from sklearn.metrics import classification_report

conf_mat = confusion_matrix(y_true=y_target, y_pred=y_simulado_lista)
print('Matriz de ConfusiГіn - DATOS ORIGINALES:\n', conf_mat)
print('MГ©tricas de Matriz de ConfusiГіn - DATOS ORIGINALES:\n',classification_report(y_target,y_simulado_lista))
labels = ['Class 0', 'Class 1']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(conf_mat, cmap=plt.cm.Blues)
fig.colorbar(cax)
ax.set_xticklabels( labels)
ax.set_yticklabels( labels)
plt.xlabel('Predicted')
plt.ylabel('Expected')
plt.show()
```
